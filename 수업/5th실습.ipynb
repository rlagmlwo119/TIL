```python
def call_10_times(func):
    for i in range(10):
        print("i =", i)
        func()
        print("=" * 20)
def print_hello():
    print("안녕하세요")
call_10_times(print_hello)
```

```python
def power(item):
    return item * item
def under_3(item):
    return item < 3
list_input = [1, 2, 3, 4, 5]
output_a = map(power, list_input)
print("map()함수의 실행결과")
print("map(power, list_input):", output_a)
print("map(power, list_input):", list(output_a))
print()

output_b = filter(under_3, list_input)
print("filter() 함수의 실행결과")
print("filter(under_3, list_input):", output_b)
print("filter(under_3, list_input):", list(output_b))
```

```python
power = lambda x : x ** 2
under_3 = lambda x : x < 3

input_list_a = [1, 2, 3, 4, 5]

output_a = map(power, input_list_a)
print("map(power, input_list_a):", output_a)
print("map(power, input_list_a):", list(output_a))
print()
output_b = filter(under_3, input_list_a)
print("filter(under_3, input_list_a):", output_b)
print("filter(under_3, input_list_a):", list(output_b))
```

```python
input_list_b = [1, 2, 3, 4, 5]
output_b = map(lambda x : x ** 2, input_list_b)
print("input_list_b):", output_b)
print("input_list_b):", list(output_b))

output_B = filter(lambda x : x < 3, input_list_b)
print("input_list_b):", output_B)
print("input_list_b):", list(output_B))
```

```python
import random
hanguls = list("가나다라마바사아자차카타파하")
with open("info.txt", "w") as file:
    for i in range(1000):
        name = random.choice(hanguls) + random.choice(hanguls)
        weight = random.randrange(40, 100)
        height = random.randrange(140, 200)
        file.write("{}, {}, {}\n".format(name, weight, height))
```

```python
with open("info.txt", "r") as file:
    for line in file:
        (name, weight, height) = line.strip().split(", ")
        if (not name) or (not weight) or (not height):
            continue
        bmi = int(weight) / ((int(height) / 100) **2)
        result = ""
        if 25 <= bmi:
            result = "과체중"
        elif 18.5 <= bmi:
            result = "정상"
        else:
            result = "저체중"
        print('\n'.join(["이름: {}",
                         "몸무게: {}",
                          "키: {}",
                          "BMI: {}",
                          "결과: {}"
                        ]).format(name, weight, height, bmi, result))
        print()
```

```python
numbers = list(range(1, 10 + 1))
print("#홀수만 추출")
print(list(filter(lambda x: x % 2 == 1, numbers)))
print("3 이상, 7 미만 추출")
print(list(filter(lambda x: 3 <= x < 7, numbers)))
print("# 제곱해서 50미만 추출")
print(list(filter(lambda x: x **2 < 50, numbers)))
```
