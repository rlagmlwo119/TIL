```python
import pandas as pd
```

```python
df = pd.read_excel("서울지역 대학교 위치.xlsx", encoding="euc-kr", index_col = 0)
```

```python
seoul_map = folium.Map(location=[37.55, 126.98],
                       zoom_start=12,
                      tiles="Stamen Terrain")
```

```python
for name, lat, lng in zip(df.index, df.위도, df.경도):
    folium.Marker([lat, lng], tooltip=name).add_to(seoul_map)
```

```python
seoul_map
```

```python
seoul_map_circle = folium.Map(location=[37.55, 126.98],
                       zoom_start=12,
                      tiles="Stamen Terrain")
```

```python
for name, lat, lng in zip(df.index, df.위도, df.경도):
    folium.CircleMarker([lat, lng], 
                       radius = 10,
                       color="blue", 
                       fill=True, 
                       fill_color = "blue", 
                       fill_opacity=0.5,
                       tooltip = name).add_to(seoul_map_circle)
```

```python
seoul_map_circle
```

```python
geo_data = json.load(open("경기도행정구역경계.json", encoding="utf-8"))
```

```python
g_map = folium.Map(location=[37.5502, 126.982],
                   zoom_start=10)
```

```python
year = 2017
```

```python
folium.Choropleth(geo_data = geo_data, 
                 data = df[year], 
                 columns=[df.index, df[year]],
                  fill_color="YlOrRd",
                  fill_opacity=0.5, 
                  line_opacity=0.2,
                  threshold_scale = [10000, 100000, 300000, 500000, 700000],
                  key_on = "feature.properties.name"
                 ).add_to(g_map)
```

```python
df_seoul = df[df["시도명"] == "서울특별시"].copy()
```
