```python
seoul_map = folium.Map(location=[37.55, 126.98], 
                      zoom_start=12,
                      tiles="Stamen Terrain")
```

```python
seoul_map
```

```python
for name, lat, lng in zip(df.index, df["위도"], df["경도"]):
    folium.Marker(location=[lat, lng], tooltip=name).add_to(seoul_map)
```

```python
seoul_map
```

```python
for name, lat, lng in zip(df.index, df["위도"], df["경도"]):
    folium.CircleMarker(location=[lat, lng],
                       radius=10,
                       color="brown").add_to(seoul_map)
```

```python
seoul_map
```

```python
seoul_map_circle = folium.Map(location=[37.55, 126.98], 
                      zoom_start=12,
                      tiles="Stamen Terrain")
```

```python
for name, lat, lng in zip(df.index, df["위도"], df["경도"]):
    folium.CircleMarker(location=[lat, lng],
                       radius=10,
                       color="brown",
                       fill=True,
                       fill_opcity=0.5,
                       fill_color="blue",
                       tooltip=name).add_to(seoul_map_circle)
```
