```python
for tr01 in table.find_all("tr"):
    print("=" * 40)
    print("tr =", tr01)
    print("tr01.find_all(td) =", tr01.find_all("td"))
#     print("tr01.find_all(td)[0] =", tr01.find_all("td")[0])
#     print("tr01.find_all(td)[1] =", tr01.find_all("td")[1])
    
    for td01 in tr01.find_all("td"):
        print("*" * 40)
        print("td01 =", tr01)
        print("td01.find(a) =", td01.find("a"))
        
        if td01.find("a"):
            print("None 아니라서 True 여기 실행")
            print("td01.find(a).text =", td01.find("a").text) # 동네이름
            print("tr01.find_all(td)[5] =", tr01.find_all("td")[5])
            print("tr01.find_all(td)[5].text =", tr01.find_all("td")[5].text) #온도
            print("tr01.find_all(td)[9] =", tr01.find_all("td")[9])
            print("tr01.find_all(td)[9].text =", tr01.find_all("td")[9].text) #습도
            data.append(
                [
                    td01.find("a").text, 
                    tr01.find_all("td")[5].text, 
                    tr01.find_all("td")[9]
                ]
            )
        else:
            print("None 이라서 False 여기 실행")
        print("*" * 40)
        
    print("=" * 40)
```

```python
table = soup.find("table", {"class": "table_develop3"})
data = []
```

```python
for tr1 in table.find_all("tr"):
    td = tr1.find_all("td")
    
    for td1 in td:
        if td1.find("a"):
            location = td1.find("a").text
            temp = td[5].text
            hum = td[9].text
            data.append([location, temp, hum])
```

```python
for i in data:
    print("i =", i)
    print("i[0] =", i[0])
    print("i[1] =", i[1])
    print("i[2] =", i[2])
    print("{1}, {0}, {2}".format(100, 200, 300)) #{}안에 숫자를 써서 나오는 순서 변경가능
    print("{0}, {1}, {2}".format(100, 200, 300))
    print("{0}, {1}, {2}".format(i[0], i[1], i[2]))
    print("=" * 50)
```

```python
with open("weather.csv", "w") as file:
    file.write("location,temperature,humidity\n")
    for i in data:
        file.write("{0},{1},{2}\n".format(i[0], i[1], i[2]))
```
