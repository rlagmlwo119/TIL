```python
f = open("subwaytime.csv")
data = csv.reader(f)
next(data)
next(data)
```

```python
mx = [0] * 24
mx_station = [""] * 24
for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):
        a = row[j*2 + 4]
        if a > mx[j]:
            mx[j] = a
            if j + 4 > 24:
                mx_station[j] = row[3] + "(" + str(j+4 - 24) + ")"
            else:
                mx_station[j] = row[3] + "(" + str(j+4) + ")"
```

```python
print(mx_station)
print(mx)
```

```python
f = open("subwaytime.csv")
data = csv.reader(f)
next(data)
next(data)
s_in = [0] * 24
s_out = [0] * 24
for row in data:
    row[4:] = map(int, row[4:])
    for i in range(24):
        s_in[i] = s_in[i] + row[4+i*2]
        s_out[i] = s_out[i] + row[5+i*2]
```

```python
time = range(4, 28)
plt.figure(dpi = 300)
plt.plot(s_in, label="승차")
plt.plot(s_out, label="하차")
plt.xticks(range(24), time)
plt.legend
```
