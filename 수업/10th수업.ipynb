```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome("chromedriver.exe")
```

```python
driver.get("https://www.youtube.com/channel/UCyn-K7rZLXjGl7VXGweIlcA/videos")
```

```python
e1 = driver.find_element_by_tag_name("body")
```

```python
from selenium.webdriver.common.keys import Keys
import time
```

```python
for index in range(10):
    e1.send_keys(Keys.END)
    time.sleep(5)
```

```python
driver.execute_script("alert(document.documentElement.scrollHeight)") #현재 화면의 길이
```

```python
last_height = driver.execute_script("return document.documentElement.scrollHeight")

last_height
```

```python
while True:
    last_height = last_height = driver.execute_script("return document.documentElement.scrollHeight")
    print("스크롤 전 화면길이 =", last_height)
    for i in range(10):
        e1.send_keys(Keys.END)
        time.sleep(3)
        
    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    print("스크롤 10번 후 화면길이 =", new_height)
    if new_height == last_height:
        print("끝까지 스크롤 됨")
        break
```

```python
page = driver.page_source
```

```python
from bs4 import BeautifulSoup as BS
```

```python
page_html = BS(page, "html.parser")
```

```python
all_videos = page_html.find_all(id="details")
```

```python
all_videos
```

```python
title_list = []
for element in all_videos:
    title = element.find(id="video-title")
    print("title:", title.text)
    if len(title.text.strip()) > 0:
        title_list.append(title.text)
```

```python
count_list = []
for element in all_videos:
    count_num = element.find(("span", {"class": "style-scope ytd-grid-video-renderer"}))
    print("count_num:", count_num.text)
    count_list.append(count_num.text)
```

```python
count_num_list = []
for number in count_list:
    number = number.replace("조회수", "")
    num = float(number[:-2])
    danwee = number[-2:]
    if danwee == "만회":
        count_num_list.append(int(num*10000))
```

```python
count_num_list
```

```python
dic_youtube = {"title": title_list, "view_num": count_num_list}
```

```python
import pandas as pd
youtube = pd.DataFrame(dic_youtube)
youtube
```
