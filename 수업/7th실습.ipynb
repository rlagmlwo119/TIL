```python
앉힐수있는최대사람수 = 10
앉힐수있는최소사람수 = 2
전체사람수 = 6
def 문제(남은사람수, 앉힌사람수):
    print("=" * 30)
    print("문제({}.{})시작".format(남은사람수, 앉힌사람수))
    counter = 0
    if 남은사람수 == 0:
        #global counter
        #counter = counter + 1
        print("끝, 1 리턴")
        return 1
    if 남은사람수 < 0:
        print("음수, 0 리턴")
        return 0
    if 남은사람수 == 1:
        print("한명, 0 리턴")
        return 0
    
    for i in range(앉힌사람수, 앉힐수있는최대사람수 + 1):
        print("i(앉힌사람수)=", i)
        print("남은사람수 - i=", 남은사람수 - i)
        counter = counter + 문제(남은사람수 - i, i)
        print("문제({}, {})가 끝. counter = {}".format(남은사람수 - i, i, counter))
    return counter
        
```

```python
앉힐수있는최대사람수 = 10
앉힐수있는최소사람수 = 2
전체사람수 = 100
memo = {}

def 문제(남은사람수, 앉힌사람수):
    print("=" * 30)
    print("문제({}.{})시작".format(남은사람수, 앉힌사람수))
    key = str([남은사람수, 앉힌사람수])
    print("key =", key)
    if key in memo:
        print("=" * 30)
        print("memo에 있음, counter 리턴", memo[key])
        return memo[key]
    
    counter = 0
    if 남은사람수 == 0:
        #global counter
        #counter = counter + 1
        print("끝, 1 리턴")
        return 1
    if 남은사람수 < 0:
        print("음수, 0 리턴")
        return 0
    if 남은사람수 == 1:
        print("한명, 0 리턴")
        return 0
    
    for i in range(앉힌사람수, 앉힐수있는최대사람수 + 1):
        print("i(앉힌사람수)=", i)
        print("남은사람수 - i=", 남은사람수 - i)
        counter = counter + 문제(남은사람수 - i, i)
        memo[key] = counter
        print("문제({}, {})가 끝. counter = {}".format(남은사람수 - i, i, counter))
    return counter
```

```python
앉힐수있는최소사람수 = 2
앉힐수있는최대사람수 = 10
전체사람수 = 6
counter = 0
def 문제(남은사람수, 앉힌사람수):
    print("문제({}, {})시작".format(남은사람수, 앉힌사람수))
    for i in range(앉힌사람수, 앉힐수있는최대사람수 + 1):
        print("i(앉힌사람수) =", i)
        print("남은사람수 - i :", 남은사람수 - i)
        if 남은사람수 - i == 0:
            print("다 앉음")
            global counter
            counter = counter + 1
```
