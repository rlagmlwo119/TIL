```python
def factorial(n):
    output = 1
    for i in range(1, n + 1):
        output = output * i
    return output
print(factorial(1))
print(factorial(2))
print(factorial(5))
print(factorial(10))
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
print(factorial(1))
print(factorial(2))
print(factorial(5))
print(factorial(10))
```

```python
def fibonacci(n):
    if n == 1:
        return 1
    if n == 2:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
```

```python
counter = 0
def fibonacci(n):
    print("fibonacci({})를 구함".format(n))
    global counter
    counter = counter + 1
    if n == 1:
        return 1
    if n == 2:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
print(fibonacci(15))
print("fibonacci(15) 계산에 활용된 덧셈횟수는 {}번임".format(counter))
```

```python
counter = 0
dictionary = {1: 1, 2: 2}
def fibonacci(n):
    global counter
    counter = counter + 1
    if n in dictionary:
        return dictionary[n]
    else:
        output = fibonacci(n - 1) + fibonacci(n - 2)
        dictionary[n] = output
        return output

print(fibonacci(15))
print(counter)
```
