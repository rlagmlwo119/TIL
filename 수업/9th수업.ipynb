```python
from selenium import webdriver as wd
url = "http://tour.interpark.com"
```

```python
driver = wd.Chrome("chromedriver.exe")
```

```python
driver.get(url)
```

```python
element = driver.find_element_by_id("SearchGNBText")
```

```python
keyword = "도쿄"
```

```python
element.send_keys(keyword)
```

```python
driver.find_element_by_class_name("search-btn").click()
```

```python
driver.find_element_by_id("li_R").click()
```

```python
driver.execute_script("alert('hi')")
```

```python
driver.execute_script("searchModule.SetCategoryList(3,'')")
```

```python
for page in range(1, 5):
    print("searchModule.SetCategoryList({},'')".format(page))
```

```python
for page in range(1, 5):
    driver.execute_script("searchModule.SetCategoryList({},'')".format(page))
    time.sleep(10)
```

```python
boxitem = driver.find_elements_by_css_selector(".panelZone>.oTravelBox>.boxList>li")
```

```python
boxitem.get_attribute("outerHTML")
```

```python
for li in boxitem:
    print(li.find_element_by_css_selector("h5.proTit").text)
    print(li.find_element_by_css_selector(".proPrice").text.split("원")[0])
```

```python
for page in range(1, 5):
    print("page =", page)
    #print("=" * 80)
    driver.execute_script("searchModule.SetCategoryList({}, '')".format(page))
    #print("=" * 80)
    time.sleep(5)
    boxitem = driver.find_elements_by_css_selector(".panelZone>.oTravelBox>.boxList>li")
    print("=" * 80)
    
    for li in boxitem:
        print(li.find_element_by_css_selector("h5.proTit").text)
        print(li.find_element_by_css_selector(".proPrice").text.split("원")[0])
        print("=" * 80)
```
