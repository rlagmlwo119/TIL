```python
import requests
from bs4 import BeautifulSoup

response = requests.get("http://www.weather.go.kr/weather/observation/currentweather.jsp")
soup = BeautifulSoup(response.text, "html.parser")

table = soup.find("table", {"class": "table_develop3"})
data = []
for tr_1 in table.find_all("tr"):
    td_1 = tr_1.find_all("td")
    
    for td_2 in td_1:
        if td_2.find("a"):
            place = td_2.find("a").text
            temperature = td_1[5].text
            humidity = td_1[9].text
            data.append([place, temperature, humidity])
```

```python
data
```

```python
with open("weather_16.csv", "w") as file:
    file.write("place,temperature,humidity\n")
    for i in data:
        file.write("{},{},{}\n".format(i[0], i[1], i[2]))
```

```python
response = requests.get("https://sonyunara.com/shop/list.php?page=1&cate=0104&limit=60")
soup = BeautifulSoup(response.text, "html.parser")
```

```python
soup
```

```python
category_box = soup.find("div", {"class": "category-product"})
category_box
```

```python
category_box.find_all("div", {"class": "item-content"})
```

```python
data = []
for item in category_box.find_all("div", {"class": "item-content"}):
    name = item.find("div", {"class": "subject"}).find("a").text
    print("name =", name)
    print("=" * 80)
    price = item.find("div", {"class": "pull-left"}).find_all("span")[1].text
    print("price =", price)
    print("=" * 80)
    data.append([name, price])
```

```python
data
```

```python
with open("shop.csv", "w") as file:
    file.write("name,price\n")
    for item in data:
        file.write("{},{}\n".format(item[0], item[1]))
```
