```python
from selenium import webdriver as wd
url = "http://tour.interpark.com"
```

```python
driver = wd.Chrome("chromedriver.exe")
```

```python
driver.get(url)
```

```python
element = driver.find_element_by_id("SearchGNBText")
```

```python
element.send_keys("도쿄")
```

```python
driver.find_element_by_class_name("search-btn").click()
```

```python
driver.find_element_by_id("li_R").click()
```

```python
driver.execute_script("searchModule.SetCategoryList(2, '')")
```

```python
for n in range(1, 5):
    print("n =", n)
    driver.execute_script("searchModule.SetCategoryList({}, '')".format(n))
    print("time stop")
    time.sleep(8)
    print("{}번째 끝".format(n))
```

```python
driver.execute_script("searchModule.SetCategoryList(1, '')")
```

```python
boxitem = driver.find_elements_by_css_selector(".panelZone>.oTravelBox>.boxList>li")
```

```python
for n in range(1, 5):
    print("page =", n)
    print("=" * 80)
    driver.execute_script("searchModule.SetCategoryList({}, '')".format(n))
    time.sleep(5)
    boxitem = driver.find_elements_by_css_selector(".panelZone>.oTravelBox>.boxList>li")
    
    for list_box in boxitem:
        print("상품:", list_box.find_element_by_css_selector("h5.proTit").text)
        price = list_box.find_element_by_css_selector(".proPrice").text.replace("~","")
        print("가격:", price)
        print("=" * 80)
```
